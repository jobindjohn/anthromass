## Hand measurements:

**Hand breadth:** he breadth of the right hand between the landmarks at metacarpale II and metacarpale V

**Hand circumference:** The circumference of the right hand

**Hand length:** The length of the right hand between the stylion landmark on the wrist and the tip of the middle finger


```python
hand = ['handbreadth','handcircumference','handlength']
```

## Scree plot:

```python
scree_plot(ansur_scaled,hand,'Hand')
```

The plot shows that the PC1 explains 98.1%, PC2 expalins 1.9% and PC3 0.1% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,hand,'Hand','Gender')
```

The biplot shows the most influencing variables. It also shows that the **Hand lenght** and the **Hand circumference** are the variables standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,hand,'Hand',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
