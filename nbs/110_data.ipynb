```python
#| export
import pandas as pd
from sklearn.model_selection import train_test_split
```

### Normalize
One function for normalizing one measurement, it is normalized based on the ansur data. 

```python
#| export
def normalize(data:pd.DataFrame, measurement:str):
    ansur=pd.read_csv('../data/processed/ANSURIImalefemale.csv')
    avg_m = ansur[measurement].mean()
    std_m = ansur[measurement].std()
    data[measurement] = (data[measurement] - avg_m) / std_m
    return data
```

One function for normalizing multiple measurements.

```python
#| export
def normalize_all(data:pd.DataFrame, measurement:list):
    for m in measurement:
        data = normalize(data, m)
    return data
```

The function below substract the mean of ansur measurement.

```python
#| export
def minus_mean(data:pd.DataFrame, measurement:list):
    ansur=pd.read_csv('../data/processed/ANSURIImalefemale.csv')
    datacopy = data.copy()
    for m in measurement:
        avg_m = ansur[m].mean()
        datacopy[m] = datacopy[m].astype(float) - avg_m
    return datacopy
```

Here the normalized and minus mean functions have been applied to the ansur data and new data files have been created.

```python
ansur=pd.read_csv('../data/processed/ANSURIImalefemale.csv')
normalized = normalize_all(ansur, ['weightkg', 'stature'])
normalized.to_csv('../data/processed/ANSURIInormalized.csv', index=False)
```

```python
ansur=pd.read_csv('../data/processed/ANSURIImalefemale.csv')
minus_mean=minus_mean(ansur, ['weightkg', 'stature'])
minus_mean.to_csv('../data/processed/ANSURIImalefemaleminusmeans.csv', index=False)
```

### Train and test data

This is a function for splitting the data into a test and train set of your chosing.

```python
#| export
def make_train_test(data:pd.DataFrame, test_size:float):
    train, test = train_test_split(data, test_size=test_size, shuffle=True, random_state=42)
    return train, test
```

The three code cells below generates new data files containing test and train sets.

```python
normalize=pd.read_csv('../data/processed/ANSURIInormalized.csv')
train, test = make_train_test(normalized,0.2)
train.to_csv('../data/processed/ANSURIInormalizedtrain.csv', index=False)
test.to_csv('../data/processed/ANSURIInormalizedtest.csv', index=False)
```

```python
ansur=pd.read_csv('../data/processed/ANSURIImalefemale.csv')
train, test = make_train_test(ansur, 0.2)
train.to_csv('../data/processed/ANSURIItrain.csv', index=False)
test.to_csv('../data/processed/ANSURIItest.csv', index=False)
```
