## Foot measurements:

**Foot breadth horizontal:** The maximum breadth of the right foot

**Foot length:** The maximum length of the right foot

**Ball of foot circumference:** The circumference of the foot at the first and fifth metatarsophalangeal landmarks

```python
foot = ['footbreadthhorizontal','footlength', 'balloffootcircumference']
```

## Scree plot:

```python
scree_plot(ansur_scaled,foot,'Foot')
```

The plot shows that the PC1 explains 98.5%, PC2 expalins 1.4% and PC3 0.1% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,foot,'Foot','Gender')
```

The biplot shows the most influencing variables. It also shows that the **Foot length** and **Ball of foot circumference** are the variables standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,foot,'Foot',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
