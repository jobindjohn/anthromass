## Thigh measurements:

**Thigh circumference:** "The circumference of the right thigh at the gluteal furrow landmark

**Thigh clearance:** The vertical distance between a sitting surface and the thigh point, top landmark

**Thigh link:** The difference between the trochanterion height and the lateral femoral epicondyle height 

<p style="margin-left: 40px;">
<b>Trochanterion height: </b> The vertical distance between a standing surface and the trochanterion landmark</p>

<p style="margin-left: 40px;">
<b>Lateral femoral epicondyle height: </b> The vertical distance between a
standing surface and the standing lateral femoral epicondyle landmark</p>


```python
thigh = ['thighcircumference','thighclearance','thighlink'] 
```

## Scree plot:

```python
scree_plot(ansur_scaled,thigh,'Thigh')
```

The plot shows that the PC1 explains 92.9%, PC2 expalins 6.8% and PC3 0.3% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,thigh,'Thigh','Gender')
```

The biplot shows the top most influencing variables. It also shows that the **Thigh circumference** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,thigh,'Thigh',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
