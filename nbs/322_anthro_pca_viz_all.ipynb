## All measurements:

**Bizygomatic breadth:** The maximum horizontal breadth of the face (between the zygomatic arches) at the left
and right zygion landmarks

**Head breadth:** The maximum horizontal breadth of the head above the plane of attachment of the ears

**Head circumference:** The maximum circumference of the head above the attachment of the ears

**Head length:** The distance from the glabella landmark to the opisthocranion landmark 

**Menton-sellion length:** The distance from the glabella landmark to the opisthocranion landmark 

**Shoulder height:** The difference between acromialheight and axillaheight

<p style="margin-left: 40px;">
<b>acromialheight:</b> The vertical distance between a standing surface and the right acromion landmark</p>

<p style="margin-left: 40px;">
<b>axillaheight:</b> The vertical distance between a standing surface and the anterior-scye-on-the-torso landmark</p>

**Chest circumference:** The maximum horizontal circumference of the chest at the level of chest point,

**Chest depth:** The horizontal distance between the chest point anterior landmark and the back at the same level 

**Chest breadth:** The maximum horizontal breadth of the chest at the level of the chest point anterior landmark

**Middle trunk length:** the difference between chest height and waist height omphalion

<p style="margin-left: 40px;">
<b>Chest height:</b> the vertical distance between a standing surface and the chest point anterior landmark </p>

<p style="margin-left: 40px;">
<b>Waist height omphalion:</b> The vertical distance between a standing surface and the anterior omphalion landmark </p>

**Waist breadth:** The horizontal breadth of the waist at the level of omphalion

**Waist circumference:** The horizontal circumference of the waist, passing over all omphalion landmarks

**Waist depth:** The horizontal distance between the anterior and posterior omphalion landmarks

**Lower trunk length:** the difference between waist height (omphalion) and trochanterion height

<p style="margin-left: 40px;">
<b>Waist height (omphalion): </b> The vertical distance between a standing surface and the anterior omphalion landmark </p>

<p style="margin-left: 40px;">
<b>Trochanterion height: </b> The vertical distance between a standing surface and the trochanterion landmark
</p>

**Hip breadth:** The horizontal distance between the lateral buttock landmarks

**Buttock circumference:** The horizontal circumference of the trunk at the level of the buttock point, posterior, right and left lateral landmarks

**Abdominal extension depth, sitting:** The horizontal distance between the abdominal point anterior and the back at the same level

**Bicristal breadth:** The straight-line distance between the right and left iliocristale landmarks

**Waist back length:** The surface distance between the cervicale landmark and the posterior omphalion landmark 

**Thigh circumference:** "The circumference of the right thigh at the gluteal furrow landmark

**Thigh clearance:** The vertical distance between a sitting surface and the thigh point, top landmark

**Thigh link:** The difference between the trochanterion height and the lateral femoral epicondyle height 

<p style="margin-left: 40px;">
<b>Trochanterion height: </b> The vertical distance between a standing surface and the trochanterion landmark</p>

<p style="margin-left: 40px;">
<b>Lateral femoral epicondyle height: </b> The vertical distance between a
standing surface and the standing lateral femoral epicondyle landmark</p>

**Ankle circumference:** The minimum horizontal circumference of the right ankle

**Calf link:** The difference between the lateral femoral epicondyle height and the lateral malleolus height

<p style="margin-left: 40px;">
<b>Lateral femoral epicondyle height: </b> The vertical distance between a
standing surface and the standing lateral femoral epicondyle landmark</p>

<p style="margin-left: 40px;">
<b>Lateral malleolus height: </b>
The vertical distance between a standing surface and the standing lateral femoral epicondyle landmark <p>

**Calf circumference:** The maximum horizontal circumference of the right calf

**Foot breadth horizontal:** The maximum breadth of the right foot

**Foot length:** The maximum length of the right foot

**Ball of foot circumference:** The circumference of the foot at the first and fifth metatarsophalangeal landmarks

**Hand breadth:** he breadth of the right hand between the landmarks at metacarpale II and metacarpale V

**Hand circumference:** The circumference of the right hand

**Hand length:** The length of the right hand between the stylion landmark on the wrist and the tip of the middle finger

**Biceps circumference flexed:** he circumference of the right upper arm around the flexed biceps brachii muscle at the biceps point

**Acromion radiale length:** The distance between the right acromion landmark and the radiale landmark

**Forearm circumference flexed:** The circumference of the flexed right forearm is measured with a tape passing across the crease at the juncture between the upper arm and the forearm

**Wrist circumference:** The circumference of the wrist, perpendicular to the long axis of the forearm

**Forearm length:** The difference between Forearm-hand length and the hand length

<p style="margin-left: 40px;">
<b>Forearm-hand length: </b> he horizontal distance between the olecranon rear landmark and the dactylion III landmark</p>

<p style="margin-left: 40px;">
<b>Hand length: </b> The length of the right hand between the stylion landmark on the wrist and the tip of the middle finger<p>

**Neck circumference:** The circumference of the neck at the level of the infrathy-roid landmark (Adam’s apple)

**Neck circumference base:** The circumference of the base of the neck
