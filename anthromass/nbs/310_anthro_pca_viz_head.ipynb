## Head measurements:

**Bizygomatic breadth:** The maximum horizontal breadth of the face (between the zygomatic arches) at the left
and right zygion landmarks

**Head breadth:** The maximum horizontal breadth of the head above the plane of attachment of the ears

**Head circumference:** The maximum circumference of the head above the attachment of the ears

**Head length:** The distance from the glabella landmark to the opisthocranion landmark 

**Menton-sellion length:** The distance from the glabella landmark to the opisthocranion landmark 

```python
head= ['bizygomaticbreadth','headbreadth','headcircumference','headlength','mentonsellionlength']
```

## Scree plot:

```python
scree_plot(ansur_scaled,head,'Head')
```

The plot shows that the PC1 explains 99% of the variation for the body section and is the primary PC relevant to investigate.

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,head,'Head','Gender')
```

The biplot shows that the **headcircumference** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,head,'Head',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the headcircumference is the varable with the biggetst influence in the PCA.
