## Upper arm measurements:

**Biceps circumference flexed:** he circumference of the right upper arm around the flexed biceps brachii muscle at the biceps point

**Acromion radiale length:** The distance between the right acromion landmark and the radiale landmark



```python
upper_arm = ['bicepscircumferenceflexed','acromionradialelength']
```

## Scree plot:

```python
scree_plot(ansur_scaled,upper_arm,'Upper arm')
```

The plot shows that the PC1 explains 92.4% and PC2 expalins 7.6% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,upper_arm,'Upper arm','Gender')
```

The biplot shows the most influencing variables. It also shows that the **Acromion radiale length** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,upper_arm,'Upper arm',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
