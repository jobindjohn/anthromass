## Lower trunk measurements:

**Lower trunk length:** the difference between waist height (omphalion) and trochanterion height

<p style="margin-left: 40px;">
<b>Waist height (omphalion): </b> The vertical distance between a standing surface and the anterior omphalion landmark
</p>

<p style="margin-left: 40px;">
<b>Trochanterion height: </b> The vertical distance between a standing surface and the trochanterion landmark
</p>

**Hip breadth:** The horizontal distance between the lateral buttock landmarks

**Buttock circumference:** The horizontal circumference of the trunk at the level of the buttock point, posterior, right and left lateral landmarks

```python
lower_trunk = ['hipbreadth','lowertrunklength','buttockcircumference']
```

## Scree plot:

```python
scree_plot_with_elbow(ansur_scaled,lower_trunk,'Lower trunk')
```

The plot shows that the PC1 explains 98% of the variation for the body section and is the primary PC relevant to investigate.

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,lower_trunk,'Lower trunk','Gender')
```

The biplot shows that the **Buttock circumference** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,lower_trunk,'Lower trunk',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
