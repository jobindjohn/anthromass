# Predicting Body Segment Properties

> anthromass nbdev 2

*A tool for predicting body segment properties based on the data set ANSURII*

## About 

This tool predicts body segment properties, including **mass**, **center of mass**, and **moment of inertia**, based on the input variables **height**, **mass** and **gender**. 

The goal is to enable easier estimation of biomechanical properties for use in simulations, ergonomics, or health-related analyses.

The project is built and trained on anthropometric measurements from the **ANSURII** dataset (U.S. Army Anthropometric Survey). 

When predicting measurements models were made using the libraries XGBoost and Bambi. 

The data was investigated with Principal Component Analysis (PCA). The two questions in focus of the analasis was:
* Is there a difference in measurements between genders?
* Which measurements are highly correlated with others?

The final output of combining predictions of measurements and mass- and inertial properties is both a 2D-model and a 3D-model of the predicted human. The visualizations comes together with
* posistions of center of mass and joint centers
* estimated segment masses using two different regression models
* moment of inertia for each segment

## Developer Guide

The structure of our project starts with a folder ’nbs’, short for notebooks, where all
code is written. The files are numerated, and the files for handling the data starts at
100. The regression files starts at 200, followed by the PCA at 300, and mass estimation
at 600. Finally at 1000 all different parts are put together in a function where input is
height, mass, and gender and the output is a visualization of the estimated human. The
Python modules which hold all reusable functions are generated in another folder called
antropmass. The data used for the project is located in a folder called data. This folder
includes two sub folders, one for raw data and one for processed data. Lastly, there is an
output folder where tables, figures and the models are saved.

## How to use

* Go to file [1000_final_result.ipynb]{nbs\1000_final_result.ipynb}.
* Go to the code cell where the function *make_a_human()* is called upon.
* Insert desired height, mass and gender. Where height is in mm, mass in kg, and gender is 1 if male and 0 if female
* Run

Example:

````python
make_a_human(1760,86,1)
````

* Go to *generate_urdf_files()*
* Insert desired height, mass and gender. Where height is in mm, mass in kg, and gender is 1 if male and 0 if female.
* Run
* Files are now created in the model_output folder, found in antropmass/mass.
* Reveal the model_output folder in file explorer.
* Bring all files from the folder to an urdf viewer, we have been using this one: https://gkjohnson.github.io/urdf-loaders/javascript/example/bundle/

Example:

````python
generate_urdf_files(1900,70,0)
````
