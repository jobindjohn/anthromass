## About 

This tool predicts body segment properties — including **mass**, **center of mass**, and **moment of inertia** — from anthropometric measurements based on the **ANSURII** dataset (U.S. Army Anthropometric Survey).

The goal is to enable easier estimation of biomechanical properties for use in simulations, ergonomics, or health-related analyses.

## Supported File Formats

* **Input:** `.csv` — containing raw anthropometric measurements.
* **Output:** `.csv` (predicted values), `.png` (optional plots), others coming soon.
