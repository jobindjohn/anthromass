## Neck measurements:

**Neck circumference:** The circumference of the neck at the level of the infrathy-roid landmark (Adam’s apple)

**Neck circumference base:** The circumference of the base of the neck



```python
neck = ['neckcircumference','neckcircumferencebase']
```

## Scree plot:

```python
scree_plot_with_elbow(ansur_scaled,neck,'Neck')
```

The plot shows that the PC1 explains 99.3% and PC2 expalins 0.7% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,neck,'Neck','Gender')
```

The biplot shows the most influencing variables. It shows that both varaibles in this sections are very similar for PC1 in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,neck,'Neck',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
