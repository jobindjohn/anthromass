## Middle trunk measurements:

**Middle trunk length:** The difference between chest height and waist height (omphalion)
    

<p style="margin-left: 40px;">
<b>Chest height:</b> The vertical distance between a standing surface and the chest point anterior landmark 
</p>

<p style="margin-left: 40px;">
<b>Waist height (omphalion):</b> The vertical distance between a standing surface and the anterior omphalion landmark </p>

**Waist breadth:** The horizontal breadth of the waist at the level of omphalion

**Waist circumference:** The horizontal circumference of the waist, passing over all omphalion landmarks

**Waist depth:** The horizontal distance between the anterior and posterior omphalion landmarks

```python
middle_trunk = ['middletrunklength','waistbreadth','waistcircumference','waistdepth']
```

## Scree plot:

```python
scree_plot(ansur_scaled,middle_trunk,'Middle trunk')
```

The plot shows that the PC1 explains 95% of the variation for the body section and is the primary PC relevant to investigate.

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,middle_trunk,'Middle trunk','Gender')
```

The biplot shows that the **waistcircumference** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,middle_trunk,'Middle trunk',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the waistcircumference is the varable with the biggetst influence in the PCA.
