```python
import pandas as pd
import bambi as bmb
from anthropmass.anthro_module import *
```

```python
train=pd.read_csv('../data/processed/ANSURIInormalizedtrain.csv')
```

```python
def make_model(measurement, data):
    variables = f"{measurement} ~ weightkg + stature + Gender"
    made_model = bmb.Model(variables,data=data)
    made_fitted = made_model.fit(tune=2000, draws=2000, init="adapt_diag", chains=4, cores=1, progressbar=False)
    return made_fitted
```

### Here are the pickles made, obs only mac users 

```python
y_train=train.iloc[:,1:94].drop('weightkg',axis=1).drop('stature',axis=1)
variables = y_train.columns[0:2]
kindofmodel = 'bambi'
for var in variables:
    model = make_model(var, train)
    make_pickle(model, kindofmodel, var)
```
