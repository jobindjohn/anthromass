## Head measurements:

**bizygomaticbreadth:** The maximum horizontal breadth of the face (between the zygomatic arches) at the left
and right zygion landmarks

**headbreadth:** The maximum horizontal breadth of the head above the plane of attachment of the ears

**headcircumference:** The maximum circumference of the head above the attachment of the ears

**headlength:** The distance from the glabella landmark to the opisthocranion landmark 

**mentonsellionlength:** The distance from the glabella landmark to the opisthocranion landmark 

```python
head= ['bizygomaticbreadth','headbreadth','headcircumference','headlength','mentonsellionlength']
```

```python
scree_plot(ansur_scaled,head,'Head')
```
