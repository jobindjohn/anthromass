## Shank measurements:

**Ankle circumference:** The minimum horizontal circumference of the right ankle

**Calf link:** The difference between the lateral femoral epicondyle height and the lateral malleolus height

<p style="margin-left: 40px;">
<b>Lateral femoral epicondyle height: </b> The vertical distance between a
standing surface and the standing lateral femoral epicondyle landmark</p>

<p style="margin-left: 40px;">
<b>Lateral malleolus height: </b>
The vertical distance between a standing surface and the standing lateral femoral epicondyle landmark <p>

**Calf circumference:** The maximum horizontal circumference of the right calf


```python
shank = ['anklecircumference','calflink','calfcircumference']
```

## Scree plot:

```python
scree_plot(ansur_scaled,shank,'Shank')
```

The plot shows that the PC1 explains 95.4%, PC2 expalins 4.1% and PC3 0.5% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,shank,'Shank','Gender')
```

The biplot shows the most influencing variables. It also shows that the **Calf link** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,shank,'Shank',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
