Here the models are made and fitted with the function "fit_model" which can be found in 201_anthropometry.ipynb. The formula can be changed and kindofmodel is want we decided to name these models, this can therefore also be changed when trying a new formula. 

```python
variables = measurement_names()
kindofmodel = 'bambi_c'
formula='0 + C(Gender) + Component + weightkg + stature'
for var in variables:
    model = fit_model(var, formula)
    make_pickle(model, kindofmodel, var)
```
