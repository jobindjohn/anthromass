## Forearm measurements:

**Forearm circumference flexed:** The circumference of the flexed right forearm is measured with a tape passing across the crease at the juncture between the upper arm and the forearm

**Wrist circumference:** The circumference of the wrist, perpendicular to the long axis of the forearm

**Forearm length:** The difference between Forearm-hand length and the hand length

<p style="margin-left: 40px;">
<b>Forearm-hand length: </b> he horizontal distance between the olecranon rear landmark and the dactylion III landmark</p>

<p style="margin-left: 40px;">
<b>Hand length: </b> The length of the right hand between the stylion landmark on the wrist and the tip of the middle finger<p>



```python
forearm = ['forearmcircumferenceflexed','wristcircumference','forearmlength']
```

## Scree plot:

```python
scree_plot_with_elbow(ansur_scaled,forearm,'Forearm')
```

The plot shows that the PC1 explains 95.8%, PC2 expalins 3.6% and PC3 0.6% of the variation for all body sections thogether

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,forearm,'Forearm','Gender')
```

The biplot shows the most influencing variables. It also shows that the **Forearm length** is the variable standing out the most in this body section

## Plot of Loadings:

```python
plot_loadings(ansur_scaled,forearm,'Forearm',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the buttock circumference is the varable with the biggetst influence in the PCA.
