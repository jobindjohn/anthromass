## Read the file

```python
import pandas as pd
```

In the function below the ansur data for male and female is combined. The column subjectid for females had a capital letter, which was changed to a lowercase letter so that this columns for male and female could be combined. The values in the column weightkg was in the raw data expressed in hg therefore the column was divided with 10 to correct the misstake. 


```python
def read_ansur_raw_file():
    datamale=pd.read_csv('../data/raw/ANSURIIMALEPublic.csv', encoding='latin')  
    datafemale=pd.read_csv('../data/raw/ANSURIIFEMALEPublic.csv')  

    datafemale = datafemale.rename(columns= {"SubjectId":"subjectid"}) 
    ansur = pd.concat([datamale, datafemale])
    
    ansur["weightkg"] = ansur["weightkg"]/10
    return ansur
```

In this function below the genders was change from male and female to 1 and 0.

```python
def female_male():  
    ansur=read_ansur_raw_file()
    ansur['Gender']=ansur['Gender'].map({'Male': 1, 'Female': 0})
    return ansur
```

In this function a data file is generated where males and females are joined in one combined datafile, the males are represented by the number 1 and females are represented by the number 0. This data file also includes the changes for weightkg that were mentioned above. 

```python
ansur = female_male()
ansur.to_csv('../data/processed/ANSURIImalefemale.csv', index=False)
```
