## Upper trunk measurements:

**shoulderheight:** the difference between acromialheight and axillaheight

<p style="margin-left: 40px;">
<b>acromialheight:</b> The vertical distance between a standing surface and the right acromion landmark</p>

<p style="margin-left: 40px;">
<b>axillaheight:</b> The vertical distance between a standing surface and the anterior-scye-on-the-torso
landmark</p>

**chestcircumference:** The maximum horizontal circumference of the chest at the level of chest point,

**chestdepth:** The horizontal distance between the chest point anterior landmark and the back at the
same level 

**chestbreadth:** The maximum horizontal breadth of the chest at the level of the chest point anterior
landmark

```python
upper_trunk = ['shoulderheight', 'chestcircumference','chestdepth','chestbreadth']
```

## Scree plot:

```python
scree_plot_with_elbow(ansur_scaled,upper_trunk,'Upper trunk')
```

The plot shows that the PC1 explains 98% of the variation for the body section and is the primary PC relevant to investigate.

## Biplot:

```python
PC2_plot_biplot(ansur,ansur_scaled,upper_trunk,'Upper trunk','Gender')
```

The biplot shows that the **chestcircumference** is the variable standing out the most in this body section

```python
plot_loadings(ansur_scaled,upper_trunk,'Upper trunk',1)
```

Since the PC1 was the primary relevant PC the plot shows only the loadings for PC1. The plot shows, in line with the biplot, that the chestcircumference is the varable with the biggetst influence in the PCA.
